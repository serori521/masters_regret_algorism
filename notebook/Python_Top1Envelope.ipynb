# Python Top-1 Envelope Demo

```python
import numpy as np
from python.top1_envelope import track_top1_envelope
n = 5; TL, TU = 0.05, 0.40; T0 = TU
rng = np.random.default_rng(0)
cuts = np.array([TU, 0.32, 0.24, 0.16, TL]); segs = len(cuts)-1
A_segments = [rng.normal(0.0, 3.0, size=n) for _ in range(segs)]
B_segments = [rng.normal(0.0, 1.0, size=n) for _ in range(segs)]
def seg_index(T):
    for k in range(segs):
        if cuts[k+1] <= T <= cuts[k]: return k
    return segs-1
def update_coefficients(T):
    k = seg_index(T); return A_segments[k], B_segments[k]
def linear_horizon(T, A, B):
    k = seg_index(T); return T - cuts[k+1]
out = track_top1_envelope(T0, TL, TU, n, update_coefficients, linear_horizon)
out
```
